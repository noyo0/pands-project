```python
import pandas as pd
SOURCEDATA="iris.data"
#----read in with giving headers to each column-----------------------
headers=[
    "sepal length (cm)", 
    "sepal width (cm)", 
    "petal length (cm)", 
    "petal width (cm)",
    "species"]
df=pd.read_csv(SOURCEDATA, names=headers)
print(df)
```

### relation between data

```python
import seaborn as sns
ax_x = "sepal length (cm)"
ax_y = "sepal width (cm)"
sns.relplot(x=ax_x, y=ax_y, hue='species', data=df) # sns.relplot = relation plot
plt.suptitle('sepal width and lenght relations within species')
```

### Univariate plotting
##### Univariate analysis explores each variable in a data set, separately. It looks at the range of values, as well as the central tendency of the values. It describes the pattern of response to the variable. It describes each variable on its own. Descriptive statistics describe and summarize data.

##### Relationship of the data with the target

+ using a 'target' column with identical values for each entries per species we can visualise each attribute separately.

```python
# set target value 0,1,2 for each species ref: https://datagy.io/pandas-conditional-column/
df.loc[df['species']== 'Iris-setosa','target']=0
df.loc[df['species']== 'Iris-versicolor','target']=1
df.loc[df['species']== 'Iris-virginica','target']=2
```

#### Show data on scatter plot separated by target values 0,1 and 2

```python
ax_x = "sepal length (cm)"
ax_y = "target"
sns.relplot(x=ax_x, y=ax_y, hue='species', data=df) # sns.relplot = relation plot
plt.suptitle(f'{ax_x}, comparison',y=1.07)
```
